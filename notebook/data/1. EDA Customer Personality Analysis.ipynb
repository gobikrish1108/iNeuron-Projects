#### 2.2 Dataset information

### Dataset Column Descriptions

- **ID**: Unique identifier for each customer.
- **Year_Birth**: Birth year of the customer.
- **Education**: Level of education of the customer.
- **Marital_Status**: Marital status of the customer.
- **Income**: Annual income of the customer.
- **Kidhome**: Number of small children in customer's household.
- **Teenhome**: Number of teenagers in customer's household.
- **Dt_Customer**: Date when the customer was enrolled.
- **Recency**: Number of days since the last purchase.
- **MntWines**: Amount spent on wine products.
- **MntFruits**: Amount spent on fruit products.
- **MntMeatProducts**: Amount spent on meat products.
- **MntFishProducts**: Amount spent on fish products.
- **MntSweetProducts**: Amount spent on sweets products.
- **MntGoldProds**: Amount spent on gold products.
- **NumDealsPurchases**: Number of purchases made with a discount.
- **NumWebPurchases**: Number of purchases made through the website.
- **NumCatalogPurchases**: Number of purchases made through catalog mailings.
- **NumStorePurchases**: Number of purchases made directly in stores.
- **NumWebVisitsMonth**: Number of visits to the website in the last month.
- **AcceptedCmp3, AcceptedCmp4, AcceptedCmp5, AcceptedCmp1, AcceptedCmp2**: Customer responses to different marketing campaigns (binary: 1 if customer accepted the offer in the campaign, 0 otherwise).
- **Complain**: Whether the customer has complained (binary: 1 if customer complained in the last 2 years, 0 otherwise).
- **Z_CostContact**: Constant variable with a value of 3.
- **Z_Revenue**: Constant variable with a value of 11.
- **Response**: Customer response to the last campaign (binary: 1 if customer accepted the offer in the last campaign, 0 otherwise).

### 4. Data Preprocessing

##### In this stage of data processing, our goal is to determine the age of each customer using their birth year. Given that customers registered with the company between 2012 and 2014, and for simplicity, we assume data was collected in January 2015. This involves deriving a new feature that reflects the age of customers at the time of data capture.

```python
import datetime as dt
df['Age'] = 2015 - df.Year_Birth
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'],format='%d-%m-%Y')
df['Customer_Tenure_Month'] = 12.0 * (2015 - df.Dt_Customer.dt.year ) + (1 - df.Dt_Customer.dt.month)
```

### 4.1 Feature Engineering

#### Total spending

```python
df['Total_Expenses'] =  df.MntWines + df.MntFruits + df.MntMeatProducts + df.MntFishProducts + df.MntSweetProducts + df.MntGoldProds
```

#### Age groups

```python
df.loc[(df['Age'] >= 13) & (df['Age'] <= 19), 'AgeGroup'] = 'Teen'
df.loc[(df['Age'] >= 20) & (df['Age']<= 39), 'AgeGroup'] = 'Adult'
df.loc[(df['Age'] >= 40) & (df['Age'] <= 59), 'AgeGroup'] = 'Middle Age Adult'
df.loc[(df['Age'] > 60), 'AgeGroup'] = 'Senior Adult'
```

#### Number of children

```python
df['Children'] = df['Kidhome'] +df['Teenhome']
```

#### Marital Status

```python
df.Marital_Status = df.Marital_Status.replace({'Together': 'Married',
                                                           'Married': 'Married',
                                                           'Divorced': 'Single',
                                                           'Widow': 'Single', 
                                                           'Alone': 'Single',
                                                           'Absurd': 'Single',
                                                           'YOLO': 'Single'})
```
